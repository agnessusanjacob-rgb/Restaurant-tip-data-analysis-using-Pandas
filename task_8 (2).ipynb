## 19. Data Visualization using pandas

```python
#smoker count using bar chart
tips_data['smoker'].value_counts().plot(kind='bar')

```

**Insight:**
1. The number of smokers is much lower than non-smokers.
2. The restaurant is more popular among people who don't smoke.


```python
#day piechart
tips_data['day'].value_counts().plot(kind='pie')
```

**Insight:**
1. Saturday is the busiest day of the week.
2. Friday is the least busy day with the smallest share of visits

```python
#days vs tips using  line chart
# using reindex to get the days in correct order (Thur -> Sun)
tips_data.groupby('day')['tip'].mean().reindex(['Thur', 'Fri', 'Sat', 'Sun']).plot(kind='line', marker='o', title='Average Tip by Day')
```

**Insight:**
1. Sunday shows the highest average tip amount.
2. There is a steady growth in tipping behavior from Thursday to Sunday.

```python
# Scatter plot  for amount and tip
tips_data.plot(kind='scatter', x='amount', y='tip', title='Amount vs Tip Amount', color='teal', alpha=0.6)
```

**Insight:**
1. There is a clear positive relationship: as the bill amount goes up, the tip also increases.
2. A few "generous" outliers exist where small bills resulted in high tips.


```python
# gender and day count using stacked bar chart
pd.crosstab(tips_data['day'], tips_data['sex']).reindex(['Thur', 'Fri', 'Sat', 'Sun']).plot(kind='bar', stacked=True, color=['pink', 'skyblue'], title='Gender Distribution by Day')
```

**Insight:**
1. Men outnumber women on every single day.
2. The gender gap is largest on Saturday and Sunday.
3. Friday has the smallest number of total customers but the most even balance between men and women.
